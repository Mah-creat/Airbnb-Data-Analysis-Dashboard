```python
import pandas as pd
import numpy as np
```

```python
chicago_listings = pd.read_csv(r"E:\PowerBI_Projects\Airbnb_Dashboard_Project\data\chicago_listings.csv")
chicago_reviews = pd.read_csv(r"E:\PowerBI_Projects\Airbnb_Dashboard_Project\data\chicago_reviews.csv")

new_orleans_listings = pd.read_csv(r"E:\PowerBI_Projects\Airbnb_Dashboard_Project\data\new_orleans_listings.csv")
new_orleans_reviews = pd.read_csv(r"E:\PowerBI_Projects\Airbnb_Dashboard_Project\data\new_orleans_reviews.csv")
```

```python
# Remove neighbourhood_group column
chicago_listings = chicago_listings.drop(columns=['neighbourhood_group'])

# Fill missing host names
chicago_listings['host_name'] = chicago_listings['host_name'].fillna("Unknown")

# Replace missing reviews_per_month with 0
chicago_listings['reviews_per_month'] = chicago_listings['reviews_per_month'].fillna(0)

# Replace missing last_review
chicago_listings['last_review'] = chicago_listings['last_review'].fillna("No Review")

# Fill missing price with median
chicago_listings['price'] = chicago_listings['price'].fillna(chicago_listings['price'].median())

# Replace missing license
chicago_listings['license'] = chicago_listings['license'].fillna("Not Available")
```

```python
new_orleans_listings = new_orleans_listings.drop(columns=['neighbourhood_group'], errors='ignore')
```

```python
new_orleans_listings['host_name'] = new_orleans_listings['host_name'].fillna("Unknown")

new_orleans_listings['reviews_per_month'] = new_orleans_listings['reviews_per_month'].fillna(0)

new_orleans_listings['last_review'] = new_orleans_listings['last_review'].fillna("No Review")

new_orleans_listings['price'] = new_orleans_listings['price'].fillna(new_orleans_listings['price'].median())

new_orleans_listings['license'] = new_orleans_listings['license'].fillna("Not Available")
```

```python
new_orleans_listings.isnull().sum()
```

```python
chicago_listings['city'] = "Chicago"
new_orleans_listings['city'] = "New Orleans"
```

```python
import pandas as pd

combined_listings = pd.concat([chicago_listings, new_orleans_listings])
```
